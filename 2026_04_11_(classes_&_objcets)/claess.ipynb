```python
# How we could acsses a variable of a class from outside of the class:-
class My_class:
    a = 200
    b = 400
c1 = My_class
print(c1.a)
```

```python
class My_class:
    def My_method(self):
        self.a = 250
        self.b = 350
        self.c = self.a+self.b
        print(f"a = {self.a}")
        print(f"b = {self.b}")
        print(f"c = {self.c}")
        return self.c
c2 = My_class()
c2.My_method()
c2.a = c2.a + 200
print(f"a = {c2.a}")
print(c2.My_method())
```
