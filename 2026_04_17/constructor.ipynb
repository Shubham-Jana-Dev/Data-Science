```python
#3 convert Kelvin to F
# formula ==>> f = (k-273)x(9/5)+32
def temp():
    k  = int(input("Enter the temparature in K:- "))
    f = (k - 273)*(9/5)+32
    print(f"{k} K = {f} F")
temp()
```

```python
#4 Convert C to K
def temparature():
    C = float(input("Enter the temparature in C:- "))
    kelvine = C + 273.15
    print(f"{C} C = {kelvine} K")
temparature()
```

```python
#5. convert KM to Nano Meter
def distance():
    km = float(input("Enter the distance in KM:- "))
    nano = km*10**(12)
    print(f"{km} km = {nano} nano meter")
distance()
```

```python
#6. convert miles to pico meter
def distance():
    m = float(input("Enter the distance in miles:- "))
    p = m*(10**15)
    print(f"{m} miles = {p} picometer.")
distance()
```

```python
#8
def resistance():
    r1 = float(input("Enter the value of the first resistanc:- "))
    r2 = float(input("Enter the value of the second resistanc:- "))
    r3 = float(input("Enter the value of the third resistanc:- "))
    sirice = r1 + r2 + r3
    paralal = (1/r1) + (1/r2) + (1/r3)
    print(f"The total resistance for series connection = {sirice}\nThe total resistance for paralal connection = {paralal}")
resistance()
```

```python
def area_cylinder():
    r = float(input("Enter the redious of the rod:- "))
    h = float(input("Enter the hight of the rod:- "))
    area = ((2*3.14*(r**2))+(2*3.14*r*h))
    print("The area of the rod is ",area)
area_cylinder()
```
