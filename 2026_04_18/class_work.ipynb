```python
def myFunc(*args, **kwargs):
    print("Non-Keyword Argument (*args):")
    for arg in args:
        print(arg)

    print('\nKeyword Arguments (**kwags):')
    for key, value in kwargs.items():
           print(f"{key} == {value}")

myFunc('Hi',"Welcome",first = 'Shubham\'s',mid='for',last='Books')
```

```python
def Add(a,b):
    return a+b
def subtact(a,b):
    return a - b
def multiply(a,b):
    return a * b
def divide(a,b):
    return a/b if b != 0 else "Zero divition error"
def square(n):
    return n**2
def cube(n):
    return n**3
def is_even(n):
    return n%2 == 0 
def is_odd(n):
    return n%2 != 0
def max_of_two(a,b):
    return a if a > b else b
def max_of_three(a,b,c):
    return max(a,b,c)
```

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact *= i
    return fact
def fibonacci(n):
    a, b = 0,1
    result = []
    for _ in range(n):
        result.append(a)
        a,b = b, a+b
def is_prinme(n):
    if n < 2:
        return False
    return True
def sum_of_digits(n):
    return sum(int(d) for d in str(n))
def reverse_number(n):
    return int(str(n)[::-1])
def is_palindrome_number(n):
    return str(n) == str(n)[::-1]

def gcd(a,b):
    while b:
        a,b = b,a%b
    return a
def lcm(a,b):
    return a * b // gcd(a,b)
def power(base,exp):
    return base ** exp
def amstrong(n):
    digits = [int(d) for d in str(n)]
    power = len(digits)
    return n == sum(d**power for d in digits)

```
